```python
# Escribe tu código aquí para explorar el dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import os
import urllib.request
import tarfile
from pathlib import Path
```

```python
data_url = "https://github.com/ageron/data/raw/main/housing.tgz"
data_dir = Path('../data/raw')
data_dir.mkdir(parents=True, exist_ok=True)
file_path = data_dir / "file.tar.gz"
```

```python

urllib.request.urlretrieve(data_url, file_path)
with tarfile.open(file_path, "r:gz") as tar:
  tar.extractall(data_dir)
print(f"Data downloaded and extracted to {data_dir}")
```

```python
df = pd.read_csv(data_dir / 'housing' / 'housing.csv')
df.head()
```
